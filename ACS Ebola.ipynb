Census Data for Understanding West-African Immigrant Populations
=============
In reading about Ebola, I noticed a dearth of information about immigrant communities from the countries afflicted by the disease. The [American Community Survey](http://www2.census.gov/acs2013_1yr/summaryfile/Sequence_Number_and_Table_Number_Lookup.xls) includes information about ancestry and place of birth for Americans. This seemed like a natural place to start looking for more hard figures about these communities. Though the data is not especially user-friendly, by digging into the [documentation](http://www2.census.gov/acs2013_1yr/summaryfile/ACS_2013_SF_Tech_Doc.pdf), one can find such information broken out by ancestry and place of birth.

This notebook, which should be fully reproducible for other users (though probably not Windows users), describes how to get the relevant data, and presents some shallow analysis of the demographics of US cities.

```python
#Download the data for the 2013 (most recent) ACS
!wget http://www2.census.gov/acs2013_1yr/summaryfile/2013_ACSSF_All_In_1_Giant_File\(Experienced-Users-Only\)/All_Geographies.zip
!unzip All_Geographies.zip -d All_Geographies
```

Finding the data
----------------
This ACS data has many files which correspond to a large range of questions. The two questions in particular that we look at here, are

1. FIRST ANCESTRY REPORTED
2. PLACE OF BIRTH FOR THE FOREIGN-BORN POPULATION IN THE UNITED STATES

[This excel sheet](http://www2.census.gov/acs2013_1yr/summaryfile/Sequence_Number_and_Table_Number_Lookup.xls) allows us to find the table and column information for the data we're looking for. Although Guinea is the largest country currently experiencing a significant outbreak of Ebola, it is not represented among the answers to the ACS, so is not included in our analysis.

```python
#Specify location of data we're interested in.
offset = 5 #Number of id columns before data.
#Columns for ancestry.
sierra_leone = 81 + offset
liberia = 78 + offset
#Columns for place of birth
pob_l = 111 + offset
pob_s = 113 + offset
```

```python
#Read in the data for ancestry.
state='us' #Use the national file
df = pd.read_csv('All_Geographies/e20131%s0007000.txt' % state, header=None,
                 dtype={3:str, 4:str, 5:str})
#Read in the data for place of birth
pob = pd.read_csv('All_Geographies/e20131%s0015000.txt' % state, header=None,
                 dtype={3:str, 4:str, 5:str})
```

```python
#Uncomment these lines to download data for header information for geo files.
!wget http://www2.census.gov/acs2013_1yr/summaryfile/UserTools/2013_Summary_FileTemplates.zip
!unzip 2013_Summary_FileTemplates.zip 2013_SFGeoFileTemplate.xls
```

```python
#Get the column names for the geo files.
geo_names = pd.read_excel('2013_SFGeoFileTemplate.xls')
mgeo = pd.read_csv('All_Geographies/g20131%s.csv' % state,
                   names=geo_names.columns,
                   header=None,
                   dtype={'LOGRECNO':str,
                          'SUMLEVEL':str})
```

```python
def add_place(df, sierra_leone, liberia):
    """This function merges on location data
    to the another table with LOGRECNO.
    """
    merge = pd.merge(df[[5, sierra_leone, liberia]], 
                     mgeo[['LOGRECNO', 'NAME', 'SUMLEVEL']], 
                     left_on=5, right_on='LOGRECNO')
    merge.rename(columns={sierra_leone:'sierra_leone',liberia:'liberia'}, inplace=True)
    #merge = merge[merge.SUMLEVEL=='060']
    merge = merge[merge[['liberia', 'sierra_leone']].sum(axis=1) > 0]
    return merge
merge = add_place(df, sierra_leone, liberia)
merge_pob = add_place(pob, pob_s, pob_l)
```

```python
def make_city(merge):
    """Makes a barplot after some data cleanup
    """
    #Only consider the metropolitan level data (others are possible)
    city = merge[merge.SUMLEVEL=='400']
    city['NAME'] = city.NAME.apply(lambda x: x[0:-21])
    p = city.sort('liberia', ascending=False)
    return p.set_index('NAME')[['liberia', 'sierra_leone']]
```

Results
==============

Having loaded this data, we can now see which metropolitan areas have the largest populations from Sierra Leone and Liberia. Because people in these communities are more likely to travel to and from Liberia, these are places where healthcare workers, CDC resources etc. ought to be especially focused. The charts below show the estimated (by the Census Bureau) numbers of people in each area reporting Liberia or Sierra Leone as their place of ancestry (first chart), or birth (second chart).

```python
c = make_city(merge)
c.plot(kind='bar', figsize=(16,10), title="Reported as Ancestry")
```

```python
c_pob = make_city(merge_pob)
c_pob.plot(kind='bar', figsize=(16,10), title="Reported as Place of Birth")
```

In [this article](http://www.latimes.com/nation/la-na-dallas-liberians-ebola-blame-20141002-story.html#page=1), the LA Times reports the size of the Liberian community in Dallas as being about 10,000 people. According to the ACS data, this would appear to be on the high side. That figure would place the Dallas-area among the top 3 largest Liberian communities, when it appears to be substantially smaller than that.

The Dallas area is however high among a very limited number of places with significant Liberian populations. Out of the estimated 80,000 people reporting Liberia as their place of birth in the US, more than half of them live in just 8 cities, while the rest are dispersed in tiny pockets.

```python
#Show the cumulative percent of total Liberian population.
total_liberian = merge_pob.liberia[0]
(c_pob.liberia / total_liberian * 100).cumsum().head(8)
```

The table below shows the ratio of the place of birth numbers to ancestry numbers. In general, because I expect immigration from Africa is a relatively recent phenomenon, I wouldn't expect the place-of-birth to ancestry ratio to deviate too much from one. While in many cases, the values are around one as expected, in Dallas 12-times as many people report being born in Liberia as having Liberian ancestry. This could reflect internal African migrations, maybe people born in Liberia maintain ancestry from some other place, thought it's not obvious why this would affect only Dallas.

On the other end, Atlanta shows more people reporting Liberian ancestry than place of birth, which could mean the immigrant population there is less recent than in other places.

```python
#Calculate the ratio of place of birth to ancestry.
r = c_pob/c
#Sort by size of liberian population to highglight most sigificant cities.
r = r.ix[c_pob.index] 
#Show only cities where there are values.
r = r[r.any(axis=1)]
r.plot(kind='bar', figsize=(16,10), title="Ratio of reported place of birth to reported ancestry")
```

Communication over Fear
-----------------
The immigrant communities shown here are already experiencing the trauma of watching the relatives and loved ones suffer from afar. No doubt they are as scared as anyone about the prospects of facing this horrible disease in the US. Despite this, there are also [reports](http://mobile.nytimes.com/2014/10/03/us/countering-fear-of-ebola-with-education-where-west-africans-live-in-us.html) that they have experienced some stigma because of their relationship to countries afflicted by the disease. One lesson learned from the the Dallas Ebola diagnosis is that effective and timely communication between patients, doctors, epidemiologists etc. is extremely important to mitigating the risks of isolated occurances of the disease. In the areas mentioned above especially, extra efforts ought to be made to ensure that communication channels (between doctors and patients, nurses and doctors, and between doctors and government) are functioning well, so that responses to any suspected cases can be as effective as possible.
